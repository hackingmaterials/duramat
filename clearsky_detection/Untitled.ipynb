```python
import pandas as pd
import numpy as np

import visualize_plotly as visualize
import pvlib
import cs_detection
```

```python
data = cs_detection.ClearskyDetection.read_srrl_dir('./srrl_cloud', 'MST', keepers=['GHI', 'Total Cloud Cover [%]', 'Opaque Cloud Cover [%]'])
data.df['Total Cloud Cover [%]'] = data.df['Total Cloud Cover [%]'].apply(lambda x: (x >= 1) * x)
data.df['Opaque Cloud Cover [%]'] = data.df['Opaque Cloud Cover [%]'].apply(lambda x: (x >= 1) * x)
```

```python
data.df.head()
```

```python
data.trim_dates('02-01-2006', '02-08-2006')
```

```python
vis = visualize.Visualizer()
vis.add_line_ser(data.df['GHI'], 'GHI')
vis.add_line_ser(data.df['Total Cloud Cover [%]'], 'TCC')
vis.add_line_ser(data.df['Opaque Cloud Cover [%]'], 'OCC')
vis.show()
```
